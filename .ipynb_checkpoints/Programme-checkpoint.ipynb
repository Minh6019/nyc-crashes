```python
import numpy as np
import pandas as pd
```

```python
url='datasets/data_100000.csv'
df_origin = pd.read_csv(url, sep=',')
#make a copy
df =df_origin.copy()
pd.set_option('display.max_columns', None)
df.head()
df.shape
```

```python
df.info() #100000 x 29 columns
```

```python
#STEP 1: Checking missing values ("" or null),

#total_na_values = df.isna().sum().sum()
#print(total_na_values)# 871235
#Sum of na values
print(f'Sum of na values by columns: \n{df.isna().sum()} \n')
```

```python
'''depend on this result: proposition to delete the colums (>90 % with non values):
 columns_to_delete = ['contributing_factor_vehicle_3',
                        'contributing_factor_vehicle_4',
                        'contributing_factor_vehicle_5',
                        'vehicle_type_code_3',
                        'vehicle_type_code_4',
                        'cross_street_name', # > 74%
                        'vehicle_type_code_5']
'''
#drop columns 
df.drop(columns=['contributing_factor_vehicle_3',
                'contributing_factor_vehicle_4',
                'contributing_factor_vehicle_5',
                'vehicle_type_code_3',
                'vehicle_type_code_4',
                'cross_street_name',
                'vehicle_type_code_5',
                'location'],
                inplace=True)
df.head(2)                        
```

```python
#STEP2 : Drop rows witn 'all' na
#df=df.dropna()
df = df.dropna(how='all')
df.shape
```

```python
#STEP3 : Drop duplicates
# a. Check for unique of data (cols)
df.duplicated()
print("Number Of Rows In The Original DataFrame:", len(df))
print("Number Of Rows After Deduping:", len(df.drop_duplicates()))

#No duplicates
```

```python
# b. I observe that "collision_id" shoul be unique
# sorting by collision_id
df.sort_values("collision_id", inplace = True)
 
# dropping ALL duplicate values in "collision_id"
df.drop_duplicates(subset ="collision_id",
                     keep = False, inplace = True)
# displaying data
df.shape
```

```python
# c. Observe values in each column (unique):
get_column_names = df.columns
for i in range(len(get_column_names)):    
    print(f'---Count values unique from column * {df.columns[i]} * ---: \n {df[df.columns[i]].unique()} \n')
```

```python
# d. Observe values in each column (nunique):
# looking for category variables
for i in range(len(get_column_names)):
    print(f'---Nr of distinct observations (unique) of column * {df.columns[i]} * is: {df[df.columns[i]].nunique()}')
```

```python
# e. Observe value_counts in each column:
get_column_names = df.columns
for i in range(len(get_column_names)):
    print(f'---Count values from column * {df.columns[i]} * --- : \n {df[df.columns[i]].value_counts()} \n')
```

```python
# step 4 : a. strip + lowewcase string columns objects:

get_column_names = df.columns

# liste of columns with string:
columns_str = []
for col in get_column_names:    
    if df[col].dtype == object:
        columns_str.append(col)
#strip + lower:        
for col in columns_str:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].astype(str).str.lower()

#pd.set_option('display.max_columns', None)
df.head()
```

```python
# step 4 : b. convert date_time
df['crash_date'] = pd.to_datetime(df['crash_date'])
df['crash_time'] = df['crash_time'].apply(lambda x: pd.Timestamp(x).strftime('%H:%M'))

df.head()
#df.info()
```

```python
# Step 5 - Replacing missing values with np.nan
 
df = df.replace('',np.nan)
print(f'All values na in the dataset: {df.isnull().sum().sum()}')
df[df.isna()].head(5)
```

```python
df = df.replace('NaT',np.nan)
df[df.isna()].head(5)

#drop NaN all rows ???
#df=df.dropna()
#df = df.dropna(how='all')
#df.shape
```

```python
#write dataFrame to csv: 1st version
df.to_csv('data_clean_GOOD_ENOUGH.csv')
```
